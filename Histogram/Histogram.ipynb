```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/AI/Data Analysis/Histogram/stdmark.csv")

dfcorrel=df[["Math","Phy","Lang"]]

correlations = dfcorrel.corr(method='pearson')

correlations
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

ax = fig.add_subplot(1,1,1)

cax = ax.matshow(correlations, vmin=-1, vmax=1)

fig.colorbar(cax)

ticks = [0,1,2]

ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(["Math","Phy","Lang"])
ax.set_yticklabels(["Math","Phy","Lang"])
plt.show()
```
